```python
#Collecting the user input
def user_pick():
    #Sssign condition and check variables
    pick = 'Wrong'
    allowed_range = range(1,10)
    in_range = False

    #Check if we have a digit and if that digit is in the range
    while pick.isdigit() == False or in_range==False:
        pick = input("Please pick a number from (1-9): ")

        #if not a digit, loop re runs until we get a digit
        if pick.isdigit() == False:
            print('Please enter a DIGIT') 

        #if digit and in allowed range, stops loop
        if pick.isdigit() == True:
            if int(pick) in allowed_range:
                in_range = True 
            else:
                print('Please pick a number from (1-9)')
                in_range = False 
    
    #return the integer version of the inputted pick
    return int(pick)


```

```python
def display_rules(row1, row2, row3): 
    #introduction to the game which explains the rules and displays the board 
    print("The following game is tic tac toe. The first to go is X and the next person is O.")
    print("Please make a selection on the board based on number location to start")
    print(row1)
    print(row2)
    print(row3)
```

```python
def display_board(row1, row2, row3):
    #Displays the current board after each turn
    print('Current Board')
    print(row1)
    print(row2)
    print(row3)
```

```python
def board_switch(board_bool):
    #Used to switch the boolean value being used to determine whose turn it is. 
    #Returns the opposite boolean value which is then sent to the board_in() function to decide if player is X or O.
    if board_bool==True:
        board_bool=False
    else:
        pass
    return board_bool
```

```python
def board_in(board_bool):
    #Recieves the board_bool value and uses it to return the string X or O on either individuals turn
    board_val=''
    if board_bool==True:
        board_val='X'
    else:
        board_val='O'
    return board_val
```

```python
def new_game():
    #Runs a new game if 'Y' string is entered
    new_game_choice = input('Would you like to play a new game? Reply with Y or N')
                         
    if new_game_choice=='Y':
        return True

```

```python

def game_win_check(board):
    #System of checks to determine if there is a winner. Could be replaced with an algorithim but for now it works.
    count = 0
    if (board[0][0]==board[0][1]==board[0][2]):
        count=3
    elif (board[1][0]==board[1][1]==board[1][2]):
        count=3
    elif (board[2][0]==board[2][1]==board[2][2]):
        count=3
    elif (board[0][0]==board[1][0]==board[2][0]):
        count=3
    elif (board[0][1]==board[1][1]==board[2][1]):
        count=3
    elif (board[0][2]==board[1][2]==board[2][2]):
        count=3
    elif (board[0][0]==board[1][1]==board[2][2]):
        count=3
    elif (board[0][2]==board[1][1]==board[2][0]):
        count=3

    else:
        pass  


    return count
```

```python
def tictactoe():
    #Set the end game condition
    #Set the board
    #Display the rules to the user
    game_on=0
    pick=0
    board=[['1' , '2' , '3' ],['4' , '5' , '6' ],['7' , '8' , '9' ]]
    row1=board[0]
    row2=board[1]
    row3=board[2]
    display_rules(row1, row2, row3)
    
    #Used to decide whose turn it is
    board_bool=False

    #used to monitor which values have already been used in the game
    allowable_values=[1,2,3,4,5,6,7,8,9]

    #While loop set up to not exceed maximum number of moves
    while game_on<9:
        
        #Sends board to game_win_check() to see if game has been won
        if game_win_check(board)==3:
            print('Congratulations ' + board_in(board_bool)+' Player. You win!!')
            break
        
        #If game hase not been won, the next user choice is selected
        choice = user_pick()

        #Checks to see if the previously selected board value has already been used
        if choice not in allowable_values:
            print("sorry that value was already used")
            
        else:
            #game_on sets up the progression of the loop
            game_on=game_on+1
            
            #changes input boolean which selects the input value. Not really necessary here but Im going to leave it as a learning moment for me.
            board_bool = not board_bool
            
            #Filters the user inputed choice and puts the necessary player mark on the tile according to the current board_bool value
            if (choice/3)<=1:
                if (choice%3)==1:
                    board[0][0]=board_in(board_bool)
        
                elif (choice%3)==2:
                    board[0][1]=board_in(board_bool) 

                else:
                    board[0][2]=board_in(board_bool)  

            elif (choice/3)>=1 and (choice/3)<=2:
                if ((choice-3)%3)==1:
                    board[1][0]=board_in(board_bool)

                elif ((choice-3)%3)==2:
                    board[1][1]=board_in(board_bool) 

                else:
                    board[1][2]=board_in(board_bool)

            else: 
                if ((choice-6)%3)==1:
                    board[2][0]=board_in(board_bool)

                elif ((choice-6)%3)==2:
                    board[2][1]=board_in(board_bool)

                else:
                    board[2][2]=board_in(board_bool)
            
            #removes the value that was just used from the list so it can not be called again
            allowable_values.remove(choice)
        
        #displays the board after the placement of the player mark
        display_board(row1, row2, row3)
    
    #Calls tictactoe() if player wants to play again
    if new_game()==True:
        tictactoe()


    

    
```
